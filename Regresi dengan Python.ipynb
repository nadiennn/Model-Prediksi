```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# Hasilkan data acak
np.random.seed(0)
x = 2.5 * np.random.randn(100) + 1.5
res = 0.5 * np.random.randn(100)
y = 2 + 0.3 * x + res

# Buat DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Tampilkan 5 baris pertama
df.head()

```

```python
# Hitunglah rata-rata x dan y
xmean = np.mean(x)
ymean = np.mean(y)

#Hitunglah suku yang dibutuhkan untuk pembilang dan penyebut beta
df['xycov'] = (df['x'] - xmean) * (df['y'] - ymean)
df['xvar'] = (df['x'] - xmean)**2

# Hitung beta dan alfa
beta = df['xycov'].sum() / df['xvar'].sum()
alfa = ymean - (beta * xmean)
print(f'alfa = {alfa}')
print(f'beta = {beta}')

ypred = alfa + beta * x # fungsi prediksi

# plot regresi terhadap data aktual
plt.figure(figsize=(12, 6))
plt.plot(x, ypred) # garis regresi
plt.plot(x, y, 'ro') # diagram sebar yang menunjukkan data aktual
plt.title('aktual vs prediksi')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
